## Answering analysis questions 

```python
Concat_df['count'] = 1
Released_Year = Concat_df.groupby(['Released_Year']).sum().reset_index()[['Released_Year','count']]
```

```python
Released_Year
```

```python
Concat_df['count'] = 1
Languages_spoken = Concat_df.groupby(['Languages spoken']).sum().reset_index()[['Languages spoken','count']]
```
